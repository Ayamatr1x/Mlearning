```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
##loading dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_datast = load_breast_cancer()
```

```python
print(cancer_datast.DESCR)
```

```python
df = pd.DataFrame(cancer_datast.data, columns=cancer_datast.feature_names)  
```

```python
df.head()
```

```python
##standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
print(scaled_data)
```

```python
###applying PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)##to extract 2 principal components
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.plot(data_pca[:,0], data_pca[:,1],
         'o',  # 'o' means circle markers
         c='orange', alpha=0.7)
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Breast Cancer Dataset (using plt.plot)')
plt.show()
```
